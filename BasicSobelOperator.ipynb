```python
from PIL import Image
from IPython.display import display
import math
path = "/workspace/Data/Lena.jpeg" # Your image path 
img = Image.open(path)
# display(img)
```

```python
def VerticalEdgeDecetor(img):
    width, height=img.size
    newimg = Image.new("RGB", (width, height), "white")
    
    for x in range(1, width-1):
        for y in range(1, height-1):
    
            Gx = 0
    
            p = img.getpixel((x, y+1))
            r = p[0]
            g = p[1]
            b = p[2]
    
            Gx += -1 * (r + g + b)
    
            p = img.getpixel((x, y-1))
            r = p[0]
            g = p[1]
            b = p[2]
    
            Gx += 1 * (r + g + b)
    
            newimg.putpixel((x,y),(Gx,Gx,Gx))
    return newimg
```

```python
def HorizontalEdgeDecetor(img):
    width, height=img.size
    newimg = Image.new("RGB", (width, height), "white")
    
    for x in range(1, width-1):
        for y in range(1, height-1):
    
            Gy = 0
    
            p = img.getpixel((x-1, y))
            r = p[0]
            g = p[1]
            b = p[2]
    
            Gy += -1 * (r + g + b)
    
            p = img.getpixel((x+1, y))
            r = p[0]
            g = p[1]
            b = p[2]
    
            Gy += 1 * (r + g + b)
    
            newimg.putpixel((x,y),(Gy,Gy,Gy))
    return newimg
```

```python
display(img)
display(VerticalEdgeDecetor(img))
display(HorizontalEdgeDecetor(img))
```
